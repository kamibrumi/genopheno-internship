# OPTION 1

```python
import networkx as nx
G = nx.Graph()
```

In order to add color, you can create an artificial attribute called color and assign the preferred color.

```python
G.add_nodes_from([
    (1, {"color": "#9a031e"}),
    (2, {"color": "#fb8b24"}),
    (3, {"color": "#fb8b24"}),
    (4, {"color": "#9a031e"}),
    (5, {"color": "#5f0f40"})
])
```

```python
G.add_edges_from([(1, 2), (1, 3), (2, 4), (2, 5)])
```

```python
G.nodes.data()
```

```python
values = [node[1]['color'] for node in G.nodes.data()]
values
```

for node in G.nodes.data():
    print(node)

```python
import matplotlib.pyplot as plt

plt.subplot(121)
nx.draw(G, with_labels=True, font_weight='bold', node_color=values, font_color='white')
plt.tight_layout()

plt.subplot(122)
nx.draw_shell(G, nlist=[range(5, 10), range(5)], with_labels=True, font_weight='bold', node_color=values, font_color='white')
plt.show()

```

# OPTION 2

You can directly just say that we want to color by a certain attribute. Say we have a graph where we have an attribute called rating:

```python
G = nx.Graph()
G.add_nodes_from([
    (1, {"rating": 1}),
    (2, {"rating": 3}),
    (3, {"rating": 4}),
    (4, {"rating": 2}),
    (5, {"rating": 4})
])

G.add_edges_from([(1, 2), (1, 3), (2, 4), (2, 5)])
ratings = [node[1]['rating'] for node in G.nodes.data()]
ratings
```

```python
plt.subplot(121)
nx.draw(G, with_labels=True, font_weight='bold', node_color=ratings, font_color='white')
plt.tight_layout()

plt.subplot(122)
nx.draw_shell(G, nlist=[range(5, 10), range(5)], with_labels=True, font_weight='bold', node_color=ratings, font_color='white')
plt.show()
```

As you can see, nodes 5 and 3 have the same color because they have the same rating. You can assign a custom color to each node by just mapping ratings to colors and giving that to the node_color parameter of the nx.draw method.
